```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('melb_data.csv')
print("Shape: ", df.shape)
df.head(10)
```

```python
df.isnull().sum()
```

```python
if 'BuildingArea' in df.columns:
  df.drop(['Unnamed: 0', 'BuildingArea', 'YearBuilt', 'CouncilArea'], axis = 'columns', inplace = True)

print("Shape: ", df.shape)
df.head(10)
```

```python
df.isnull().sum()
```

```python
ncols = []
for col in df.columns:
  if df[col].dtype in ['int64', 'float64']:
    ncols.append(col)

print(ncols)
```

```python
df_mean_impute = df.copy(deep = True)

print(ncols)
for ncol in ncols:
  df_mean_impute[ncol].fillna(df_mean_impute[ncol].mean(), inplace = True)

df_mean_impute.isnull().sum()
```

```python
df_median_impute = df.copy(deep = True)

print(ncols)
for ncol in ncols:
  df_median_impute[ncol].fillna(df_mean_impute[ncol].median(), inplace = True)

df_median_impute.isnull().sum()
```

```python
df_mode_impute = df.copy(deep = True)

for ncol in ncols:
  u = df_mode_impute[ncol].mode()
  df_mode_impute[ncol].fillna(df_mode_impute[ncol].mode().mean(), inplace = True)

df_mode_impute.isnull().sum()
```
