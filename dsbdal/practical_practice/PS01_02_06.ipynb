```python
# 1. Import all the required Python Libraries.
import pandas as pd
import sklearn.datasets
```

```python
# 2. Load the Dataset into pandas’ data frame.
from sklearn.datasets import load_iris
iris_data = load_iris()

df_iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df_iris['target'] = pd.Series(iris_data.target)
df_iris['target'].replace([0, 1, 2], iris_data.target_names, inplace=True)

df_iris

# iris_data
```

4. Turn categorical variables into quantitative variables in Python.

Method 1: Label Encoding

Method 2: One-hot Encoding

```python
# Label Encoding
df_iris_label = df_iris.copy(deep=True)

categories = df_iris['target'].unique()
categories_label = [i for i in range(len(categories))]

df_iris_label['target'].replace(categories, categories_label, inplace = True)

df_iris_label
```

```python
df_iris_onehot = df_iris.copy(deep=True)

dummies = pd.get_dummies(df_iris['target'])

dummies

df_iris_onehot = pd.concat([df_iris_onehot, dummies], axis='columns')

df_iris_onehot = df_iris_onehot.drop('target', axis='columns')

df_iris_onehot
```
