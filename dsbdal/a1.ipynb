```python
df.drop(columns = ['BuildingArea', 'YearBuilt'])
```

```python
# imputation
ncols = []
for col in df.columns:
    if(df[col].dtype in ['int64', 'float64']):
        ncols.append(col)
  
for col in ncols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
df["Price"] = df["Price"].astype('int64')
df
```

```python
scaler = MinMaxScaler()

ncols = []
for col in df.columns:
    if(df[col].dtype in ['int64', 'float64']):
        ncols.append(col)
df[ncols] = scaler.fit_transform(df[ncols])
 
df
```

```python
df_type = pd.get_dummies(df['Type'])
df_method = pd.get_dummies(df['Method'])


df_new = pd.concat([df, df_type, df_method], axis=1)
df_new
```
