```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
sns.lmplot(data=df, x="Age", y="Fare", col="Sex", hue="Pclass")
```

```python
sns.lmplot(data=df, x="Age", y="Fare", col="Survived", hue="Pclass")
```

```python
sns.displot(data=df, x="Fare", kde=True).set(title="Fare vs Count Distribution")
```

```python
df.describe()
```

```python
sns.catplot(data=df, x="Survived", y="Age")
```

```python
# survived vs gender, embarked, class

print(df['Embarked'].value_counts())

print(df['Survived'].value_counts())

print(df['Sex'].value_counts())
```

```python
sns.boxplot(data = df, x = 'Age').set_title("Age boxplot")
```

```python
cols = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
for col in cols:
    varcnt = df[col].value_counts()
    plt.bar(varcnt.index, varcnt)
    plt.title(f"{col}")
    plt.show()
```
